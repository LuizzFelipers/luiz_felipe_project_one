Extraindo tabelas de um arquivo pdf

Vou utilizar os demonstrativos de 2024 da WEG, onde podemos encontrar várias tabelas com informção importantes sobre o desempenho da empresa no último ano.


```python
import PyPDF2 
import tabula
import re
```

```python
pdf_file = open("Demonstrações Financeiras 2024.pdf",'rb')

pdf = PyPDF2.PdfReader(pdf_file)

pages = len(pdf.pages)

print(f'Quantidade de páginas: {pages}')

```

```python
page_6 = pdf.pages[6]

page_6.extract_text()
```

```python
text_format = ''.join(page_6.extract_text())

print(text_format)
```

```python
tabelas =  tabula.read_pdf("Demonstrações Financeiras 2024.pdf",
                                                  pages= 7)

```

```python
for tabela in tabelas:
    display(tabela)
    tabela = tabela.rename(columns={'Unnamed: 0':'Indicadores'})

display(tabela)
```

```python
# todas as tabelas do pdf

def visualizar_tabelas(arquivo_pdf):

    pdf_file = open(arquivo_pdf,'rb')
    
    pdf = PyPDF2.PdfReader(pdf_file)

    pages = len(pdf.pages)

    print(f'Número de páginas {pages}')

    tabelas =  tabula.read_pdf(arquivo_pdf,pages= 'all')

    for tabela in tabelas:
        display(tabela)

    
```

```python
visualizar_tabelas("Demonstrações Financeiras 2024.pdf")
```
